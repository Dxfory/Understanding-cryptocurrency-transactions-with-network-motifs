```python
import pandas as pd

# 读取完整 CSV
df = pd.read_csv('motif_profiles_all.csv')

# 构建新列名
new_columns = ['address'] + [f'Motif{i}' for i in range(1, 41)]
df.columns = new_columns

# 计算 motif 总和并排序
df['total'] = df.iloc[:, 1:].sum(axis=1)  # 计算除 address 外的 motif 总和
df_sorted = df.sort_values(by='total', ascending=False)

# 删除 total 列，只保留 address + 40个 motif 列
df_sorted = df_sorted.iloc[:, :41]

# 保存结果
df_sorted.to_csv('motif_profiles_sorted.csv', index=False)
```
