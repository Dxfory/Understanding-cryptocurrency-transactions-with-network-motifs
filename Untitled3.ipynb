```python
import pandas as pd

# 加载交易数据
df = pd.read_csv("processed_transactions.csv")
df["timeStamp"] = pd.to_datetime(df["timeStamp"])

# 获取所有唯一地址
unique_addresses = pd.unique(df[["from_address", "to_address"]].values.ravel())

# 创建节点 DataFrame
nodes_df = pd.DataFrame({"address": unique_addresses})

# 计算每个地址首次出现时间
from_seen = df.groupby("from_address")["timeStamp"].min()
to_seen = df.groupby("to_address")["timeStamp"].min()
first_seen = pd.concat([from_seen, to_seen]).groupby(level=0).min()

# 映射至节点表
nodes_df["timeStamp"] = nodes_df["address"].map(first_seen)

# 可选：添加占位属性（后续可更新为真实分类）
nodes_df["node_type"] = "unknown"

# 保存为 CSV 文件
nodes_df.to_csv("node_attributes.csv", index=False)
```

```python
from raphtory import Graph

graph = Graph()

# 载入边信息
graph.load_edges_from_pandas(
    df=df,
    src="from_address",
    dst="to_address",
    time="timeStamp",
    properties=["value"]
)

# 载入节点属性（已处理好的 nodes_df）
# 读取节点属性文件
nodes_df = pd.read_csv("node_attributes.csv")

# 将 timeStamp 列转为 datetime（如果尚未转）
nodes_df["timeStamp"] = pd.to_datetime(nodes_df["timeStamp"])

# 转换为 UNIX 毫秒时间戳（Raphtory 默认使用毫秒）
nodes_df["timeStamp"] = nodes_df["timeStamp"].astype("int64") // 10**6  # 转换为毫秒单位的整数

graph.load_nodes_from_pandas(
    df=nodes_df,
    id="address",
    time="timeStamp",
    constant_properties=["node_type"]
)
```

```python
from raphtory import algorithms
motif_counts = algorithms.global_temporal_three_node_motif(graph, 3600)
```
