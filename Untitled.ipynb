```python
from raphtory import algorithms
from detail_process import build_transaction_graph  # 你自己已有的函数
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from raphtory import plottingutils
# 1. 定义时间窗口
target_times = [
    "2022-07-09 18:00:00",
    "2022-07-02 09:00:00",
    "2022-07-05 09:00:00",
    "2022-07-10 18:00:00",
    "2022-07-01 20:00:00"
]
delta_seconds = 3600  # 1小时窗口

# 2. 加载并格式化 DataFrame
processed_transactions['timeStamp'] = pd.to_datetime(processed_transactions['timeStamp'])

# 3. 循环处理每个时间窗口
all_results = []

for ts_str in target_times:
    start_time = pd.to_datetime(ts_str)
    end_time = start_time + timedelta(seconds=delta_seconds)
    
    window_df = processed_transactions[(processed_transactions['timeStamp'] >= start_time) & (processed_transactions['timeStamp'] < end_time)]
    print(f"\n处理时间窗口：{start_time} 到 {end_time}，共 {len(window_df)} 条记录")

    if window_df.empty:
        print("⚠️ 当前时间窗口无数据，跳过。")
        continue

    # 4. 构建图
    window_df.to_csv("temp_window.csv", index=False)
    g = build_transaction_graph(window_df)

    # 5. 调用 local motif 分析
    motif_result = algorithms.local_temporal_three_node_motifs(g, delta_seconds)

    # 6. 对每个地址求 motif 总数并排序
    node_motif_summary = []
    for node_id, motif_vector in motif_result.items():
        motif_sum = int(np.sum(motif_vector))
        node_motif_summary.append({
            "address": node_id,
            "total_motifs": motif_sum,
            "motif_vector": motif_vector
        })

    sorted_nodes = sorted(node_motif_summary, key=lambda x: x["total_motifs"], reverse=True)

    # 7. 打印前3地址
    print(f"📌 时间段 {start_time} 的 Top 3 地址参与情况：")
    for i, entry in enumerate(sorted_nodes[:3]):
        print(f"Rank {i+1}: {entry['address']}")
        print(f"Total motifs: {entry['total_motifs']}")
        print(f"Motif vector: {entry['motif_vector']}")
        print("=" * 40)
   
    # 8.开始绘图
    # 继续在 sorted_nodes 中循环前10地址
    print(f"\n📌 时间段 {start_time} 的 Top 3 地址参与情况及绘图：")

    for idx, entry in enumerate(sorted_nodes[:3]):
        key = entry['address']
        motif_array = np.array(entry['motif_vector'])

        # 打印信息
        print(f"Rank {idx+1}: {key}")
        print(f"Total motifs: {entry['total_motifs']}")
        print(f"Motif vector: {motif_array.tolist()}")
        print("=" * 40)
        
        motif_array[31] += motif_array[24]
        motif_array[30] += motif_array[25]
        motif_array[30] += motif_array[26]
        motif_array[28] += motif_array[27]

        # 开始绘图
        fig, ax = plt.subplots()
        ax = plottingutils.global_motif_heatplot(motif_array, cmap="YlGnBu")
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.tick_params(axis='x', labelsize=10, pad=10)
        ax.set_title(f"Motif Heatmap for key {key[:8]}... (Rank {idx+1})")

        # 构造图像文件名
        out_img_name = f"motif_heatmap_{start_time.strftime('%Y%m%d_%H%M%S')}_rank{idx+1}_{key[:8]}.png"
        plt.savefig(out_img_name)
        plt.show()

    # 9.构造图像文件名
    out_img_name = f"motif_heatmap_{start_time.strftime('%Y%m%d_%H%M%S')}_rank{idx+1}_{key[:8]}.png"
    plt.savefig(out_img_name)
    plt.show()
    
    # 10. 存储结果
    df_out = pd.DataFrame([
        {"address": entry["address"], "total_motifs": entry["total_motifs"], **{f"motif_{i}": v for i, v in enumerate(entry["motif_vector"])}}
        for entry in sorted_nodes
    ])
    out_name = f"motif_local_counts_{start_time.strftime('%Y%m%d_%H%M%S')}.csv"
    df_out.to_csv(out_name, index=False)
    print(f"✅ 已保存结果到 {out_name}")

    all_results.append((start_time, df_out))

# 可选：合并所有结果
# merged_df = pd.concat([r[1] for r in all_results], keys=[r[0] for r in all_results])
# merged_df.to_csv("all_time_windows_local_motifs.csv")



```
